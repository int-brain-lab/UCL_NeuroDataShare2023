```python
# All released BWM-IBL data
# from 2022_10_bwm_release file
from brainwidemap import bwm_query
import numpy as np

# 1. load info about all released sessions from this file: 2022_10_bwm_release
bwm_df=bwm_query(one=None, alignment_resolved=True, return_details=False, freeze='2022_10_bwm_release')
bwm_df
```

```python
# let's search all sessions from cortexlab
# Example 2 using one.search
# one.search - gives a list of eids representing every session
sess = one.search(lab="cortexlab")
print('All sessions one.search =', len(sess))# this gives you all registered sessions from that lab, including behavior training

# this query only sessions from ephys recordings for brainwide from cortexlab
sess = one.search(lab="cortexlab",task_protocol='ephys',projects='ibl_neuropixel_brainwide_01')
print('Ephys IBL-BWM sessions one.search =', len(sess))#
sess

#Example 3 using one.alyx.rest

# one.alyx.rest - gives also metadata about each session
# this query all sessions from cortexlab
sess = one.alyx.rest('sessions', 'list',lab='cortexlab')
print('All sessions one.alyx.rest =', len(sess))# 

sess = one.alyx.rest('sessions', 'list', lab='cortexlab', task_protocol='ephys', projects='ibl_neuropixel_brainwide_01')
print('Ephys IBL-BWM sessions one.alyx.rest =', len(sess))# 
```

```python
# Example 4;
## searching sessions from sepcific subject
subject = 'SWC_054'

# query sessions endpoint
sessions = one.search(subject=subject)
print(f'No. of detected sessions: {len(sessions)}')

## query insertions ONE.ALYX.REST
insertions = one.alyx.rest('insertions', 'list', subject=subject)
print(f'No. of detected insertions: {len(insertions)}')

# after quering , each insertion has it is own metadata
insertions[0]
```
